```python
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('training_data.csv')
```

```python
df.head()
```

```python
dummies= pd.get_dummies(df.location)
dummies.head(3)
```

```python
df1 = pd.concat([df,dummies], axis = 'columns')
```

```python
df1.head()
```

```python
df2=df1.drop(['location','other','Unnamed: 0'], axis = 'columns')
df2.head(10)
```

```python
df2.shape
```

```python
# Creating our x variable for model building. The x variable are the independent variable.

x= df2.drop('price', axis = 'columns')
x.head()
```

```python
y=df2.price
y.head()
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x,y)
```
